```python
questions = ["1. How many people have heart disease and how many people doesn't havr heart disease?",
             "2. People of which sex has most heart disease?",
             "3. People of which sex has most chest pain?",
             "4. Are People with chest pain are most pron to have heart disease",
             "5. What is the average age of the individuals in the dataset?",
             "6. How many individuals have an electrocardiographic result (restecg) with a value of 1?",
             "7. What is the count of individuals with different levels of fasting blood sugar (fbs)?"]
questions
```

```python
#the answer of first question
df.target.value_counts()
```

```python
#plotting bar chart
df.target.value_counts().plot(kind = 'bar', color=["blue", "salmon"])
plt.title("Heart Disease Values")
plt.xlabel("1 = Heart Disease, 0 = No Heart Disease")
plt.ylabel("Amount");
```

```python
#pie chart
df.target.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.legend(["Heart Disease", " No Heart Disease"])
```

```python
#'0' represent 'No disease'
#'1' represent 'Disease'
df.sex.value_counts()
```

```python
#pie chart
df.sex.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.title('Male Female ratio')
plt.legend(["Male", "Female"])
```

```python
#2nd Answer
pd.crosstab(df.target, df.sex)
```

```python
sns.countplot(x = 'target', hue = 'sex', data = df)
plt.title("Heart Disease Frequency Foe sex")
plt.xlabel("0 = No Heart Disease, 1 = Heart Disease")
```

```python
#answer for 3rd question "People of which sex has which type of chest pain most?"
df.cp.value_counts()
```

```python
#bar chart
df.cp.value_counts().plot(kind = 'bar', color = ['lightskyblue', 'salmon', 'lightgreen', 'lightpink'])
plt.title("Chest Pain Type vs count")
plt.xlabel("Chest Pain Type")
plt.ylabel("Amount")
```

```python
pd.crosstab(df.sex, df.cp)
```

```python
#answer for 5th question "What is the averge age of individulas in the dataset?"
df['age'].mean()
```

```python
#bar chart
plt.figure(figsize=(10, 6))
plt.hist(df['age'], bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.axvline(df['age'].mean(), color='red', linestyle='dashed', linewidth=2, label='Average Age')
plt.legend()
plt.show()
```
