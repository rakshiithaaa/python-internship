```python
question = ["1. How many movies vs. TV shows are there?",
            "2. How does the number of titles added vary by day of the week",
            "3. hich rating is the most common?",
            "4. How many titles were released in the most recent year?",
            "5. How many titles have a duration of exactly 90 minutes?"]
```

```python
#answer for How many movies vs. TV shows are there?

# Count the number of movies and TV shows
movie_count = data[data['type'] == 'Movie'].shape[0]
tv_show_count = data[data['type'] == 'TV Show'].shape[0]

# Print the results
print("Number of Movies:", movie_count)
print("Number of TV Shows:", tv_show_count)
```

```python
#Bar chart to show the count of movies compared to TV shows

plt.bar(['Movie', 'TV Show'], [movie_count, tv_show_count], color=['blue', 'green'])
plt.xlabel('Type')
plt.ylabel('Count')
plt.title('Number of Movies vs. TV Shows')
plt.show()
```

```python
#answer for "How does the number of titles added vary by day of the week?""

df['day_added'] = pd.to_datetime(df['date_added']).dt.day_name()
day_counts = df['day_added'].value_counts()
print(day_counts)
```

```python
# bar chart
plt.bar(day_counts.index, day_counts.values, color=['blue', 'green', 'red', 'purple', 'orange', 'brown', 'pink'])
plt.xlabel('Day of the Week')
plt.ylabel('Number of Titles Added')
plt.title('Number of Titles Added by Day of the Week')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#answer for "Which rating is the most common?""

rating_counts = data['rating'].value_counts()
most_common_rating = rating_counts.index[0]

print("Most common rating:", most_common_rating)
plt.figure(figsize=(10, 6))
rating_counts.plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#pie chart
rating_counts = data['rating'].value_counts()
plt.figure(figsize=(10, 10))
plt.pie(rating_counts.values, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Ratings')
plt.show()
```

```python
#nswer for How many titles were released in the most recent year?

data['release_year'] = pd.to_datetime(data['release_year'], format='%Y').dt.year
most_recent_year = data['release_year'].max()
titles_in_most_recent_year = data[data['release_year'] == most_recent_year].shape[0]
print("Number of titles released in the most recent year ({}): {}".format(most_recent_year, titles_in_most_recent_year))
```

```python
#Bar chart

yearly_counts = data.groupby('release_year')['title'].count()
plt.figure(figsize=(12, 6))
colors = plt.cm.viridis(np.linspace(0, 1, len(yearly_counts)))  # Generate a range of colors
plt.bar(yearly_counts.index, yearly_counts.values, color=colors)
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.title('Number of Titles Released Each Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Answer for How many titles have a duration of exactly 90 minutes?

movies_90_min = df[(df['type'] == 'Movie') & (df['duration'].str.contains('90 min'))]
count_90_min = movies_90_min.shape[0]
print("Number of titles with a duration of exactly 90 minutes:", count_90_min)
```

```python
#Pie chart
other_movies_count = df[df['type'] == 'Movie'].shape[0] - count_90_min
labels = ['90 Minutes', 'Other Durations']
sizes = [count_90_min, other_movies_count]
colors = ['#ff9999','#66b3ff']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Proportion of Movies with 90 Minute Duration')
plt.show()
```
